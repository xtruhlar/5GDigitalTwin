```python
CSV_PATH = 'real5g_scenarios1-5.csv'

import pandas as pd

raw_df = pd.read_csv(CSV_PATH)
raw_df.head()
```

```python
# add columns application and log_type with zeros
raw_df['application'] = 0
raw_df['log_type'] = 0

raw_df.head()
```

```python
# uc1:	from 	2025-04-11 09:54:00	to	2025-04-11 10:05:00
uc1_interval = { "from": "2025-04-11 09:54:00", "to": "2025-04-11 10:05:00" }
# uc2: 	from 	2025-04-10 13:46:00	to	2025-04-10 13:57:20
uc2_interval = { "from": "2025-04-10 13:46:00", "to": "2025-04-10 13:57:20" }
# uc3 	from 	2025-04-10 13:28:00	to	2025-04-10 13:41:20
uc3_interval = { "from": "2025-04-10 13:28:00", "to": "2025-04-10 13:41:20" }
# uc4: 	from 	2025-04-10 13:13:35 	to	2025-04-10 13:23:33
uc4_interval = { "from": "2025-04-10 13:13:35", "to": "2025-04-10 13:23:33" } 
# uc5: 	from 	2025-04-10 14:05:48	to	2025-04-10 14:06:30
uc5_interval = { "from": "2025-04-10 14:05:48", "to": "2025-04-10 14:06:30" }
# uc6:	from	-			to	- 
uc6_interval = { "from": "-", "to": "-" }

# no_uc	the rest
```

```python
# compare the first line with the intervals
def compare_intervals(row, interval):
    start = pd.to_datetime(interval["from"])
    end = pd.to_datetime(interval["to"])
    timestamp = pd.to_datetime(row['timestamp'])
    if start <= timestamp <= end:
        return True
    else:   
        return False
```

```python
# apply the function to each row, if condition is met, in the column 'current_uc' insert uc{$i}

def apply_uc(row):
    if compare_intervals(row, uc1_interval):
        return 'uc1'
    elif compare_intervals(row, uc2_interval):
        return 'uc2'
    elif compare_intervals(row, uc3_interval):
        return 'uc3'
    elif compare_intervals(row, uc4_interval):
        return 'uc4'
    elif compare_intervals(row, uc5_interval):
        return 'uc5'
    else:
        return 'no_uc'
    
# apply the function to each row
raw_df['current_uc'] = raw_df.apply(apply_uc, axis=1)
# check the result
raw_df.head(1000)
```

```python
# to csv

raw_df.to_csv('real5g_scenarios1-5_with_uc.csv', index=False)
# check the result
df = pd.read_csv('real5g_scenarios1-5_with_uc.csv')
df.head(10)
```
